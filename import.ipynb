```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
import glob
import os
```

```python
def import_images(path, width, height):
    """
    This function imports multiple images from a folder resizes images
    converts images into 2D array and returns resulting 3D aray
    path = Location of folder
    width = Width of resized image
    height = Height of resized image
    """
    data = []
    files = glob.glob(path+'*.jpeg')
    for file in files:
        img = Image.open(file).convert('L')
        img = img.resize((width, height))  
        img = np.array(img)
        data.append(img)
    return np.array(data)
```

```python
# Importing images resizing to 256*256 pixels
train_normal_arr = import_images('chest_xray/train/NORMAL/', 256, 256)
train_pneumonia_arr = import_images('chest_xray/train/PNEUMONIA/', 256, 256)
test_normal_arr = import_images('chest_xray/test/NORMAL/', 256, 256)
test_pneumonia_arr = import_images('chest_xray/test/PNEUMONIA/', 256, 256)
val_pneumonia_arr = import_images('chest_xray/val/PNEUMONIA/', 256, 256)
val_normal_arr = import_images('chest_xray/val/NORMAL/', 256, 256)
```

```python
def ConvertToDataframe(numpy_arr):
    """
    This function intakes 3D numpy array and returns df
    by converting each image matrix to numpy array
    """    
    x = numpy_arr.shape[0]
    new_arr = numpy_arr.reshape(x, -1)
    df = pd.DataFrame(new_arr)
    return df
```

```python
# Converting matrix to array
train_normal_df = ConvertToDataframe(train_normal_arr)
train_pneumonia_df = ConvertToDataframe(train_pneumonia_arr)
test_normal_df = ConvertToDataframe(test_normal_arr)
test_pneumonia_df = ConvertToDataframe(test_pneumonia_arr)
val_normal_df = ConvertToDataframe(val_normal_arr)
val_pneumonia_df = ConvertToDataframe(val_pneumonia_arr)
```

```python
# Adding labels to dataset
train_normal_df['Class'] = 0 
train_normal_df['Data'] = 'train'
train_pneumonia_df['Class'] = 1
train_pneumonia_df['Data'] = 'train'
test_normal_df['Class'] = 0
test_normal_df['Data'] = 'test'
test_pneumonia_df['Class'] = 1
test_pneumonia_df['Data'] = 'test'
val_normal_df['Class'] = 0
val_normal_df['Data'] = 'val'
val_pneumonia_df['Class'] = 1
val_pneumonia_df['Data'] = 'val'
```

```python
data_Xrays = pd.concat([train_normal_df,train_pneumonia_df,test_normal_df,test_pneumonia_df,val_normal_df,val_pneumonia_df])

# Saving data to h5 file
data_Xrays.to_hdf('256.h5', key='stage', mode='w', index=False)
```
